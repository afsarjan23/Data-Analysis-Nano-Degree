```python
# import and load data
import pandas as pd
% matplotlib inline
df = pd.read_csv("powerplant_data_edited.csv")
df.info()
```

```python
df.hist(figsize = (7,7));
```

```python
df["Ambient"].hist();
```

```python
pd.plotting.scatter_matrix(df, figsize = (10,10));
```

```python
# plot relationship between temperature and electrical output
df.plot(x = "Ambient", y = "Energy Output", kind = "scatter");
```

```python
# plot distribution of humidity
df["Relative Humidity"].plot(kind="hist");
```

```python
# plot box plots for each variable
df["Relative Humidity"].plot(kind = "box");
```
