#### Karachi- Pakistan Vs Global Temperature comparison

I have used SQL for data extraction from the database, python and its libraries for data loading, data manipulation and visualization.
Moving average is calculated using rolling funtion of pandas. 
Key considerations were to upload the data into a dataframe initially, and then manipulate that dataframe to calculate the moving average. And finally visualization is done using python's  library matplotlib

**Data import**

```python
import pandas as pd
df = pd.read_csv("gd_kd2.csv")
import matplotlib.pyplot as plt
df.head()
```

**Data Visualization**

```python
df['Karachi_city_average'] = df['city_avg'].rolling(window=5).mean()
df['global_average'] = df['global_avg'].rolling(window=5).mean()
ax = plt.gca()
plt.title('Karachi-Pakistan vs Global avg Temperature', size = 20)
plt.ylabel('Temperature C')
df.plot(kind='line',x = 'year',y='Karachi_city_average',ax=ax, figsize=(10, 5))
df.plot(kind='line',x = 'year',y='global_average', color='red', ax=ax)

plt.show()
```

**Observation**

As we can see from the visualization that Temperature of Karachi is much higher than the global temperature. Global average temperature is below 12 degree Centigrade while Karachi Pakistan average temperature is above 25 degree centigrade. We can also see that there is an enormous fluctuation in Karachi’s temperature as well as global temperature in around 1820’s. Karachi’s temperature is not as smooth as global average  temperature, but the trend is same. Global weather trend is almost consistent with slight increase in temperature after 1975 and the increasing trend continues and same goes with Karachi’s weather trend. Hence, we can deduce that the world’s temperature is increasing and becoming a hotter place than previous years.
